Scraping and Cleaning Data for Seasons 2013-2017

```python
url_template = "https://www.basketball-reference.com/leagues/NBA_{year}_advanced.html"
```

```python
# create an empty DataFrame
advanced_df = pd.DataFrame()
```

```python
for year in range(2013, 2019): #for each year
    url = url_template.format(year=year)
    
    html = urlopen(url)
    soup = BeautifulSoup(html, 'html5lib')
    
    #get our player data
    data_rows = soup.findAll('tr')[2:]
    player_data = [[td.getText() for td in data_rows[i].findAll('td')]
                  for i in range(len(data_rows))]
    
    #Turn yearly data into a DataFrame
    year_df = pd.DataFrame(player_data, columns=column_headers)
    # create and insert the NBA Season_Yr column
    year_df.insert(0, 'NBA_Season', year)
    
    #Append to the big dataframe
    advanced_df = advanced_df.append(year_df, ignore_index=True)
```

```python
advanced_df.head()
```

```python
advanced_df.tail()
```

```python
#Convert data to proper data types
advanced_df = advanced_df.convert_objects(convert_numeric=True)

#Get rid of the rows full of null values
advanced_df = advanced_df[advanced_df.Player.notnull()]

#Replace NaNs with 0s
advanced_df = advanced_df.fillna(0)

#Rename columns
advanced_df.rename(columns={'WS/48':'WS_per_48'}, inplace=True)
#Change % symbol
advanced_df.columns = advanced_df.columns.str.replace('%', '_Perc')
```
